```python
#Initialize environment

import numpy as np
import pandas as pd
from ydata_profiling import ProfileReport
%matplotlib inline
from matplotlib import pyplot as plt 
```

```python
#Read in data file

df= pd.read_csv('SalesPerformance.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Perform EDA

ProfileReport(df)
```

```python
# Create bar plot

purpose = df['Call purpose']
duration = df['During call wait']

plt.bar(purpose, duration, color='skyblue')
plt.xlabel('Call Purpose')
plt.ylabel('Total Minutes')
plt.title('Call Duration')
plt.xticks(rotation=45, ha='right')  # Rotating x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Create line plot of call wait times

df.plot()
plt.title("Call Duration In Minutes")
plt.legend(["During call wait","Incoming call wait"])
```

```python
#Create scatterplot

duration = df['During call wait']
incoming = df['Incoming call wait']

plt.scatter(duration, incoming, color = 'g',s = 100) 
plt.xticks(rotation = 25) 
plt.xlabel('During call wait') 
plt.ylabel('Incoming Call Wait') 
plt.title('Call Wait Duration', fontsize = 20) 
  
plt.show() 
```

```python
# Create a histogram
values = df['During call wait']


plt.hist(values, bins=5, color='blue', edgecolor='black')  
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Wait Time During Calls')

plt.show()
```
